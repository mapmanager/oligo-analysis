## Purpose

Load a folder of raw images and look at the min/max intensity. The goal is to identify image stacks that are dim. Basically, dim stacks will result in cellpose dapi masks that are meaningless (too many).


```python
import numpy as np
import pandas as pd

from IPython.display import display, HTML

import oligoanalysis
```

```python
folderPath = '/Users/cudmore/Dropbox/data/whistler/data-oct-10/FST'
oaf = oligoanalysis.oligoAnalysisFolder(folderPath)

dfFolder = oaf.getDataFrame()
#print(dfFolder)

files = dfFolder['file'].values
dictList = []
for file in files:
    # files = dfFolder['']
    oa = oaf.getOligoAnalysis(file)
    
    dapiChannel = oa.getImageChannel(oligoanalysis.imageChannels.dapi)
    dapiMin = np.min(dapiChannel)
    dapiMax = np.max(dapiChannel)
    dapiRange = dapiMax - dapiMin
    
    cytoChannel = oa.getImageChannel(oligoanalysis.imageChannels.cyto)
    cytoMin = np.min(cytoChannel)
    cytoMax = np.max(cytoChannel)
    cytoRange = cytoMax - cytoMin
    
    oneDict = {
        'file': file,
        'dapiMin': dapiMin,
        'dapiMax': dapiMax,
        'dapiRange': dapiRange,
        'numDapiLabels': len(oa._dfLabels), # we get too many labels when image is dim
        'cytoMin': cytoMin,
        'cytoMax': cytoMax,
        'cytoRange': cytoRange,
    }
    dictList.append(oneDict)

df = pd.DataFrame(dictList)
display(df)
```
